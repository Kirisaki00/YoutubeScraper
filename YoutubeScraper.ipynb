```python
from selenium import webdriver;
from bs4 import BeautifulSoup;
from tqdm import tqdm;
import pandas as pd;
import time;
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

options = Options()
options.binary_location = r"C:\chrome-win64\chrome.exe"

# options.add_argument("--headless=new")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service(r"C:\chrome-win64\chromedriver.exe")
```

```python
browser = webdriver.Chrome(service=service, options=options)   
browser.get("https://www.youtube.com/@GeeksforGeeksVideos/videos")
```

```python
soup=BeautifulSoup(browser.page_source,"html.parser")   # You need to scroll to the bottom of the Video page of gfg channel to get all data from it;
sp=soup.find("ytd-two-column-browse-results-renderer").find_all("ytd-rich-item-renderer")
```

```python
data=[]
for sp in tqdm(soup.find("ytd-two-column-browse-results-renderer").find_all("ytd-rich-item-renderer")):
    title=sp.find("a",class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").text;
    VideoLink="https://www.youtube.com/"+sp.find("a",class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get("href");
    try:
        Views=sp.find_all("span",class_="inline-metadata-item style-scope ytd-video-meta-block")[0].text
    except:
        Views=None;
    try:
        DateTime=sp.find_all("span",class_="inline-metadata-item style-scope ytd-video-meta-block")[1].text
    except:
        DateTime=None;
    try:
        thumbnail=sp.find("yt-image",class_="style-scope ytd-thumbnail").find("img").get("src").split("?")[0]
    except:
        thumbanil=None
    data.append([title,Views,DateTime,VideoLink,thumbnail]);
```

```python
df1=pd.DataFrame(data,columns=["Title","Views","Date_Time","Video_Link","Thumbnail_Link"])
df1.head()
```
